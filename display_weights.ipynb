```python
import math
from h3 import h3

from IPython.display import display
import jupyter
import requests
import folium
import db


day = 2
hour = 1
URL = 'https://nominatim.openstreetmap.org/search.php?q=' + 'Lviv' + '+Ukraine&polygon_geojson=1&format=json'
r = requests.get(url=URL)
data = r.json()

coords = data[0]["geojson"]["coordinates"]
for coord in coords[0]:
    coord[0], coord[1] = coord[1], coord[0]

geoJson = {'type': 'Polygon',
           'coordinates': coords}

# geoJson = {'type': 'Polygon',
#  'coordinates': [[[37.813318999983238, -122.4089866999972145], [ 37.7866302000007224, -122.3805436999997056 ], [37.7198061999978478, -122.3544736999993603], [ 37.7076131999975672, -122.5123436999983966 ], [37.7835871999971715, -122.5247187000021967],  [37.8151571999998453, -122.4798767000009008]]] }

polyline = geoJson['coordinates'][0]
polyline.append(polyline[0])

lat = [p[0] for p in polyline]
lng = [p[1] for p in polyline]

m = folium.Map(location=[sum(lat) / len(lat), sum(lng) / len(lng)], zoom_start=13, tiles='cartodbpositron')
my_PolyLine = folium.PolyLine(locations=polyline, weight=3, color="blue")
m.add_child(my_PolyLine)
list_weights = []

hexagons = db.get_weight_of_date((day - 1) * 24 + hour)
print(hexagons)
polylines = []
lat = []
lng = []
for hex in hexagons:
    polygons = h3.h3_set_to_multi_polygon([hex[1]], geo_json=False)
    # flatten polygons into loops.
    outlines = [loop for polygon in polygons for loop in polygon]
    polyline = []
    polyline.append([outline + [outline[0]] for outline in outlines][0])
    polyline.append(int(hex[0]/5 + 2))
    color = 'red'
    if hex[0] < 15:
        color = 'orange'
    if hex[0] < 10:
        color = 'yellow'
    if hex[0] < 5:
        color = 'green'
    if hex[0] < 0:
        color = 'gray'
        polyline[1] = 2
    polyline.append(color)
    print(polyline)
    lat.extend(map(lambda v: v[0], polyline[0]))
    lng.extend(map(lambda v: v[1], polyline[0]))
    polylines.append(polyline)

for polyline in polylines:
    my_PolyLine = folium.PolyLine(locations=polyline[0], weight=polyline[1], color=polyline[2])
    m.add_child(my_PolyLine)
display(m)
```
